```python
# Data Analysis on SuperStore Sales 
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set(style="whitegrid")
```

```python
# 1. Data Loading 

# Load the dataset 
```

```python
df = pd.read_csv("SuperStore.csv")
df.head(10)
```

```python
print("Shape:", df.shape)
print("Columns:", df.columns)

df.info()
df.describe()
```

```python
# 2. Data Cleaning
```

```python
print("Missing values:\n", df.isnull().sum())
print(df.dtypes)
print("Duplicate rows:", df.duplicated().sum())
```

```python
# 3. Visualization
```

```python
# BAR CHART


# Total Sales per Category
sns.barplot(x='Category', y='Sales', data=df, estimator=np.sum)
plt.title('Total Sales per Category')
plt.show()

# Total Sales per Sub-Category
sns.barplot(x='Sub-Category', y='Sales', data=df, estimator=np.sum)
plt.xticks(rotation=90)
plt.title('Total Sales per Sub-Category')
plt.show()

# Total Sales per Segment
sns.barplot(x='Segment', y='Sales', data=df, estimator=np.sum)
plt.title('Total Sales per Segment')
plt.show()

# Total Sales per Region
sns.barplot(x='Region', y='Sales', data=df, estimator=np.sum)
plt.title('Total Sales per Region')
plt.show()

# Total Sales per State
state_order = df.groupby('State')['Sales'].sum().sort_values(ascending=False).index
plt.figure(figsize=(10, 12)) 
sns.barplot(y='State', x='Sales', data=df, estimator=np.sum, order=state_order)
plt.title('Total Sales per State')
plt.show()
```

```python
# Scatter plot

sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title('Sales vs Profit')
plt.show()

sns.scatterplot(x='Discount', y='Profit', data=df)
plt.title('Discount vs Profit')
plt.show()
```

```python
# Histogram

numeric_cols = ['Sales', 'Profit', 'Quantity', 'Discount']

for col in numeric_cols:
    plt.figure()
    df[col].hist(bins=20)
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Heatmap

plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 4. Outliner Detection 
```

```python
for col in ['Sales', 'Profit', 'Discount']:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f'{col} Boxplot')
    plt.show()
```
